```python
def pattern_index(sequence, pattern):
  pattern_idx = []
  n = len(sequence)
  k = len(pattern)

  for i in range(n - k + 1):
    seq2check = sequence[i : i+k]
    if seq2check == pattern:
      pattern_idx.append(i)


  return pattern_idx
```
