```python
def hamming_distance(p, q):
    distance = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            distance += 1

    return distance
```

```python
def approximate_pattern_matching(pattern, seq, d):
    position = []
    k = len(pattern)
    n = len(seq)

    for i in range(n - k + 1):
        if hamming_distance(seq[i : i+k], pattern) <= d:
            position.append(i)

    return position
```

```python
# Example usage:
pattern = 'ATTCTGGA'
seq = 'CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAAT'
d = 3
result = approximate_pattern_matching(pattern, seq, d)
print("Approximate Pattern Matching Positions:", result)
```
