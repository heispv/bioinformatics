```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def skew_min_idx(skew_list):
    if not skew_list:
        return []

    min_value = min(skew_list)
    min_indices = [index for index, value in enumerate(skew_list) if value == min_value]
    return min_indices
```

```python
def skewDiagram(seq):
    char_to_num_mapping = {'A': 0, 'T': 0, 'C': -1, 'G': 1}
    the_list = [char_to_num_mapping.get(char, None) for char in seq]

    skew_list = [0]
    jam = 0
    for num in the_list:
      if num != None:
        jam = jam + num
        skew_list.append(jam)

    min_idx = skew_min_idx(skew_list)

    plt.figure(figsize=(6, 3))
    plt.plot(skew_list)
    plt.show()

    return skew_list, min_idx
```

```python
text = open('/content/dataset_7_10 (2).txt')
seq = text.read()
text.close()
```

```python
skew_list, min_idx = skewDiagram(seq)
```
