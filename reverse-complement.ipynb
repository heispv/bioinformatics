```python
def reverseComplement(text):
  text = text.replace('A', 'a')
  text = text.replace('C', 'c')
  text = text.replace('T', 'A')
  text = text.replace('G', 'C')
  text = text.replace('a', 'T')
  text = text.replace('c', 'G')

  return text[::-1]
```
